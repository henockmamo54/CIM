```python
pos_c=[]
neg_c=[]
temp=[]

for i in range(pd_acc_class0.shape[0]):
#     pos_c.append(pd_ypredproba_all_class1.iloc[i,:])    
#     neg_c.append((1-pd_ypredproba_all_class1.iloc[i,:]))
    
    pos_c.append(pd_ypredproba_all_class1.iloc[i,:]*pd_ypredconfprob_all_class1.iloc[i,0] +
                (1-pd_ypredproba_all_class1.iloc[i,:])*pd_ypredconfprob_all_class1.iloc[i,1])
    
    neg_c.append((1-pd_ypredproba_all_class1.iloc[i,:])*pd_ypredconfprob_all_class1.iloc[i,3]+
                pd_ypredproba_all_class1.iloc[i,:]*pd_ypredconfprob_all_class1.iloc[i,2]
                )
    
    finalval=np.zeros((600,2))
    finalval[:,0]=np.sum(np.array(pos_c),axis=0)
    finalval[:,1]=np.sum(np.array(neg_c),axis=0)
    print(m.accuracy_score(ytest_class1,np.argmax(finalval,axis=1)), pd_acc_class1.iloc[i,0])
    
```
