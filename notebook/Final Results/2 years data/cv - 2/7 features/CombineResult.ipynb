```python
import numpy as np
import pandas as pd
```

```python
columns = [
    "Features",
    "Accuracy",
    "Precision",
    "Recall",
    "F1-score",
    "CV_Accuracy",
    "CV_Accuracy_std",
]
rf = pd.read_csv(" RF_Result.csv")
rf.columns = columns
rf["Model"] = "RF"

svm = pd.read_csv(" SVM_Result.csv")
svm.columns = columns
svm["Model"] = "SVM"

xgb = pd.read_csv(" XGB_Result.csv")
xgb.columns = columns
xgb["Model"] = "XGB"
```

```python
originalFeatures = [
    "FPG",
    "HbA1c",
    "Triglyceride",
    "BMI",
    "r-GTP gamma",
    "Uric acid",
    "Smoking",
    "Drinking",
    "Physical activity",
    "Family history",
    "SEX",
    "AGE",
]
```

```python
result = (
    pd.concat([rf, svm, xgb])
    .sort_values(by=["Accuracy"], ascending=False)
    .reset_index(drop=True)
)
result = result.iloc[:20, :]
result
```

```python
for i in range(result.shape[0]):
    temp = list(
        str(result.iloc[i, 0])
        .replace("[", "")
        .replace("]", "")
        .replace(" ", "")
        .split(",")
    )
    temp_changed = []
    for j in range(len(temp)):
        temp_changed.append(originalFeatures[int(temp[j])])
    result.iloc[i, 0] = str(temp_changed)
result
```

```python
result.to_csv("Final_result.csv", index=False)
```
